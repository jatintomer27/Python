### Threading

- In Threading all the programs are executed in the single core.

```python
import threading
```

```python
def test(id):
    print(f"The id of the function is {id}")
```

```python
test(10)
```

```python
test(3)
```

```python
test(1)
```

```python
# Now create the 3 threads for the same program to excecute in a single core.

thread = [threading.Thread(target=test,args=(i,)) for i in [10,3,5]]
thread
```

```python
for t in thread:
    t.start()
```

- The power of threading is shown here

```python
import time
```

```python
def test(x):
    for i in range(10):
        print(f"The value of x = {x} and value of i = {i}")
        time.sleep(1)
```

```python
# Now run the program without thread

test(10)
```

- The code above without thread takes 10 second to execute program for just 1 value

### Now Implement threading

```python
thread = [threading.Thread(target=test , args=(i,)) for i in [40,30,10,20]]
```

```python
# Now exceute the program with thread and 4 input

for t in thread:
    t.start()
```

- In threading it is not waiting that first only 1 input program is executed completely.

- Instead if the core is free then it will execute the other thread in that time.
