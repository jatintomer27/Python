```python
import os
import logging as lg
os.mkdir("logging")
os.chdir(os.getcwd()+"/logging")
class Data:
    def __init__(self,file_name,file_type,data,size):
        self.file_name = file_name
        self.file_type = file_type
        self.data = data
        self.size = size
        self.set_lg = False

    def file_open(self):
        try:
            self.log("we are going to open a file and write some content in it")
            with open(str(self.file_name)+"."+str(self.file_type),"w") as f:
                f.write(str(self.data))
                self.log(f"We have write the data in the file : {self.data}")
        except Exception as e:
            self.log(f"There is an error in opening a file : {e}")
            return f"There is some issue in opening a file: {e}"

    def file_read(self):
        try:
            self.log("we are going to open a file and read its some content")
            if os.path.exists(str(self.file_name)+"."+str(self.file_type)):
                f = open(str(self.file_name)+"."+str(self.file_type),"r")
                data = f.read()
                self.log(f"File content is : {data}")
                return data
            else:
                self.log("The file is not exist yet")
                return "File not exist yet"
        except Exception as e:
            self.log(f"There is an error in reading a file : {e}")
            return f"There is some issue in reading a file: {e}"
    
    def file_append(self,new_data):
        try:
            self.log(f"We are going to open a file to append some content in it.")
            if os.path.exists(str(self.file_name)+"."+str(self.file_type)):
                with open(str(self.file_name)+"."+str(self.file_type),"a+") as f:
                    self.log(f"We have appended the content in the file named as {str(self.file_name)}.{str(self.file_type)} and appended the following content : {new_data}")
                    f.write(str(new_data))
            else:
                self.log("The file is not exist yet")
                return "File is not exist yet"
        except Exception as e:
            self.log(f"There is an error in appending in the file : {e}")
            return f"There is some issue in appending in the file: {e}"
        
    def log(self,msg):
        if not self.set_lg:
            lg.basicConfig(filename=f"{os.getcwd()}/test.log",level=lg.INFO,format="%(asctime)s %(name)s %(levelname)s %(message)s")
            self.set_lg = True
        lg.info(str(msg))


        
```

## Encapsulation

- It is an idea which allow you to prevent any kind of direct access/modification of data.

```python
"""
No underscore : Public variable
Single underscore : Protected variable
Double underscore : Private variable
"""

"""
You can access the public method anywhere you want
"""

"""
You cannot access a protected variable outside of the module
"""

"""
You cannot access a private variable directly outside a class.
But there is a way to call the private variable from the object as well
"""
```

```python
class test:
    def __init__(self,a,b,c,d):
        self.a = a
        self._b = b
        self.__c = c
        self.d = d
    
    def custom(self,v):
        return v - self.a
    
    def custom2(self,v):
        return v - self._b
    
    def custom3(self,v):
        return v - self.__c
```

```python
obj1 = test(1,2,3,4)
```

```python
print(obj1.custom(7))
print(obj1.custom2(7))
print(obj1.custom3(7))
```

```python
# We can access the protected variable here

obj1._b
```

```python
# We can also modify the protected variable here

obj1._b =3
obj1._b
```

```python
# As we mention we cannot access a private variable directly outside a class

obj1.__c
```

```python
# There is a way to access a private variable outside a class as well

# We can access the private variable by using - _classname

# But this convention is not the standard way of access and modify the private variable.

obj1._test__c
```

- So what is the way to get, set and delete private variable outside a class ?

- Answer is @property decorator

- Now, when a user tries to modify the height attribute, @height.setter is called.

```python
class Tree:
   def __init__(self, height):
       self.__height = height

   @property
   def height_new(self):
       return self.__height

   @height_new.setter
   def height_new(self, new_height):
       if not isinstance(new_height, int):
           raise TypeError("Tree height must be an integer")
       if 0 < new_height <= 40:
           self.__height = new_height
       else:
           raise ValueError("Invalid height for a pine tree")
       
pine = Tree(33)
```

```python
# Now we can access the __height attribute with the help of height method

pine.height_new
```

```python
# Now we can also modify the __height attribute with the help of height setter method

# But with the limitations we want to apply on data

pine.height_new = 35
```

- If we want to show the value of height in a specific format then we can use @height.getter

```python
class Tree:
   def __init__(self, height):
       self.__height = height

   @property
   def height_new(self):
       return self.__height

   @height_new.setter
   def height_new(self, new_height):
       if not isinstance(new_height, int):
           raise TypeError("Tree height must be an integer")
       if 0 < new_height <= 40:
           self.__height = new_height
       else:
           raise ValueError("Invalid height for a pine tree")
       
   @height_new.getter
   def height_new(self):
       # You can return a custom version of height
       return f"This tree is {self.__height} meters"
       
pine = Tree(33)
```

```python
pine.height_new
```

- Always remember that encapsulation is a convention, not an enforced aspect of Python syntax.

## Iheritance

- Inheritance allows you to inherit the properties and methods of the parent class into its child classes

```python
class Car:
    def __init__(self,name,milage,engine_no,year):
        self.name = name
        self.milage = milage
        self.engine_no = engine_no
        self.year = year

    def print_milage(self):
        print(f"Milage of car is {self.milage}")
```

```python
class Audi(Car):

    # As soon as you create the method with the same as in parent class the parent class method will be overwrite.
    # To still use/keep the parent function we need to the call the same here.
    
    def __init__(self,name,milage,engine_no,year,model):
        Car.__init__(self,name,milage,engine_no,year)
        self.model = model
```

## Polymorphism


- In polymorphism same function will behave differently based on arguments and data type

- polymorphism means the same function name (but different signatures) being used for different types. 

- The key difference is the data types and number of arguments used in function

### Function polymorphism

- Here the same function is behaving differently based on the data type

```python
def test(a,b):
    # Polymorphism with respect to functions
    print(a+b)

test(1,2)
test("jatin ","tomer")
```

### Class Polymorphism

- In the below code we are able to achieve polymorphism with respect to classes 

- Here the syllabus method is behaving differently based on different class objects different 

```python
class data_science:
    def syllabus(self):
        print("This is the syllabus for the data science")

class web_development:
    def syllabus(self):
        print("This is the syllabus for the web development")

def class_parser(class_objs):
    for i in class_objs:
        i.syllabus()
```

```python
ds_obj = data_science()
wd_obj = web_development()
class_obj = [ds_obj,wd_obj]

class_parser(class_obj)
```

```python
class Test:
    def abc(self,a,b):
        return a+b
    
    def abc(self,a,b,c):
        return a+b+c
    
test_obj = Test()
```

```python
# If we write the method with same name it will be overridden

test_obj.abc(1,2)
```

### Class Method

- It is used to define a method that is bound to the class and not the instance of the class.

- Class methods are useful when you need to work with the class itself rather than any particular instance of it.

- Class method can modify the state of the class

```python
class pwskills:

    @classmethod
    def details(cls,name,email):
        return cls(name,email)
```

```python
class pwskills:

    def __init__(self,name,email):
        self.name = name
        self.email = email

    @classmethod
    def details(cls,name1,email1):
        return cls(name1,email1)

    def student_details(self):
        return self.name + " " + self.email
```

```python
class pwskills:

    # Class variable
    mobile_number = "7984984333"

    def __init__(self,name,email):
        self.name = name
        self.email = email

    @classmethod
    def details(cls,name1,email1):
        return cls(name1,email1)
    
    @classmethod
    def change_ph(cls,mobile_no):
        cls.mobile_number = mobile_no
    
    # Access the class variable inside the instance method
    def student_details(self):
        return self.name + " " + self.email + " " + pwskills.mobile_number
```

### Static Method

- A static method does not receive an implicit first argument (i.e., cls).

- A static method is also a method that is bound to the class and not the object of the class.

- This method canâ€™t access or modify the class state.

- Static methods know nothing about the class state. They are utility-type methods that take some parameters and work upon those parameters.

```python
class MyClass:
    def __init__(self, value):
        self.value = value

    @staticmethod
    def get_max_value(x, y):
        return max(x, y)

# Create an instance of MyClass
obj = MyClass(10)

print(MyClass.get_max_value(20, 30))  

print(obj.get_max_value(20, 30)) 
```

#### When to use the class or static method?
- We generally use the class method to create factory methods. Factory methods return class objects ( similar to a constructor ) for different use cases.

- We generally use static methods to create utility functions.
