## Multi-Processing

- In multiprocessing we want to execute multiple programs parallely in different cores.


```python
import multiprocessing
```

```python
def test():
    print("This is my multi-processing program")

if __name__ == '__main__':
    m = multiprocessing.Process(target=test)
    print("This is my main program")
    m.start()
    m.join()
```

- The main function is in itself a process.

- The test funtion is another process.

```python
test()
```
