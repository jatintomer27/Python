### File Operations

```python
# With the help of python we can access and manipulate text files.

f = open('file1.txt','w')
f.write("jatin tomer")
f.close()
```

```python
f = open('file1.txt','r')
f.read()
f.close()
```

```python
# To read and write into a file you can open a file in r+ mode.
# If you don't want to close the file every time then you can use this syntax
# In this syntax you don't need to close file every time manually.

with open('file1.txt','r+') as f:
    print(f.read())
    f.write("\nI want to become a Data Scientist")
```

```python
# read the file after close to see the changes

with open('file1.txt','r+') as f:
    print(f.read())
```

```python
"""
1. Create a directory.
2. Create 10 txt file and write some data in all txt at a time
3. Read all the data from these files and print in console.
4. Hold all the data in a list.
5. Write all the data in a new file from this list.
6. keep only new file in present directory and move rest in new directory.
"""
import os, shutil
if not os.path.exists('New directory'):
    l = []
    os.mkdir("New directory")
    os.chdir("New directory")
    for i in range(1,11):
        f = open(f"file_{i}","a+")
        f.write(f"This is file no {i}")
        f.close()
        f = open(f"file_{i}","r+")
        data = f.read()
        l.append(data)
        print(data)
    with open('final.txt','a+') as f:
        for line in l:
             f.write(f'{line}\n')
    os.mkdir("New directory 1")
    for i in range(1,11): 
        shutil.move(f"file_{i}",f"{os.getcwd()+'//New directory 1'}")
else:
    print("Operation is already done")
       
```

### Working with json file

```python
data = {
    "name":"Jatin",
    "mail_id":"jatin@gmail.com",
    "phone_no":7473498343,
    "subjects":['Data Science','Big Data',"System Design"]
}
```

- Write data in a file

```python
import json

with open("test.json",'w+') as f:
    json.dump(data,f,indent=4)
```

- Read data from a file

```python
with open("test.json",'r') as f:
    data1 = f.read()
    data2 = json.loads(data1)
    print(data2)
```

### Working with CSV files

```python
data = [['name','email','phone'],
        ['jatin','jatin@gmail.com',873493409],
        ['bhanu','bhanu@gmail.com',847343834]]
```

- Writing into the csv file

```python
import csv

with open('test1.csv','w+') as f:
    w = csv.writer(f)
    for i in data:
        w.writerow(i)
```

```python
# Same thing as above

with open('test2.csv','w+') as f:
    w = csv.writer(f)
    w.writerows(data)
```

- Read from a csv file

```python
with open('test1.csv','r') as f:
    r = csv.reader(f)
    for i in r:
        print(i)
```

### Buffered Read and Write 

- When the file size is very , very large then we use the Buffered Read and Write operation.

- We read the file in chunks and pieces

```python
import io
```

```python
with open('test1.txt','wb') as f:
    file = io.BufferedWriter(f)
    file.write(b"This is my first line i am trying to write\n")
    file.write(b"This is my second line i am trying to write")
    file.flush()
```

```python
with open('test1.txt','rb') as f:
    file = io.BufferedReader(f)
    data = file.read()
    print(data)
```
