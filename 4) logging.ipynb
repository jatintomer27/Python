```python
#  Importing the logging module to work with logging

import logging as lg
```

```python
# Creating the directory to store all the log details

import os

os.mkdir('logging')
os.chdir(os.getcwd() + '/logging')
```

```python
# Configuring the logger

"""
filename : File in which you want to store the logging info.    
level : The level till which you want to log the details.
format : In which format you want to log the message
"""

lg.basicConfig(filename='test.log',level=lg.INFO, format="%(asctime)s %(message)s")
```

```python
lg.info('this is for testing')
```

```python
# We are able to log for the levels whose value is more than we specified while configuring the logger

lg.warning("this is the first warning of my program")
lg.error("This is the first error")
```

```python
# To close the logging instance

lg.shutdown()
```

```python
import logging as lg
import os
import functools

# os.mkdir('logging')
# os.chdir(os.getcwd() + '/logging')
lg.basicConfig(filename='test2.log',level=lg.INFO)

def add_nums(*args):
    lg.info(f"The user numbers are : " + str(args))
    res = functools.reduce(lambda a,b:a+b,args)
    lg.info("The sum of all the numbers is : " + str(res))
    return res

result = add_nums(1,2,3,4,5)
with open('test2.log',"r+") as f:
    print(f.read())
```

```python
import logging as lg
import os

# os.mkdir('logging')
# os.chdir(os.getcwd() + '/logging')
lg.basicConfig(filename='test3.log',level=lg.INFO,format="%(asctime)s %(name)s %(levelname)s %(message)s")


# To print the message in the console
console_log = lg.StreamHandler()
console_log.setLevel(lg.INFO)
format = lg.Formatter("%(asctime)s %(name)s %(levelname)s %(message)s")
console_log.setFormatter(format)


# To add different-different users level (like root,user1,etc)
lg.getLogger('').addHandler(console_log)
user1 = lg.getLogger('user1_area')
user2 = lg.getLogger('user2_area')

lg.info("This is info level log")
user1.info('This is user1 info log')
user2.info('This is user2 info log')
user2.debug('This is user2 info log')
```
