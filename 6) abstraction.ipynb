## Abstraction classes

- Abstract class is a class in which one or more abstract methods are defined.

- In Python, abstract method feature is not a default feature.

- Abstraction means show only necessary features, hide internal logic.

- Abstraction is like using a remote control — you press buttons, but you don’t care how TV circuits work.

```python
import abc

class pwskills(abc.ABC):

    @abc.abstractmethod
    def student_details(self):
        # Abstract Method
        pass
    
    def student_assignment(self):
        # Concrete Method
        print("There are the assignments of the students")

    @abc.abstractmethod
    def student_marks(self):
        # Abstract Method
        pass
```

```python
class data_science(pwskills):
    def student_details(self):
        print("This is jatin tomer from data science")
```

```python
ds_obj = data_science()
```

- Abstract method of base class force its child class to write the implementation of the all  abstract methods defined in base class. 
    
- If we do not implement the abstract methods of base class in the child class then our code will give error

#### Concrete Method:

- Concrete methods are the methods defined in an abstract base class with their complete implementation. 

- Concrete methods are required to avoid reprication of code in subclasses.
