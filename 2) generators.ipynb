```python
# Let's create a simple function

def test1(n):
    """
    It takes a number and return the list of squares of number till that numbers from 0.
    """
    l = []
    for i in range(n):
        l.append(i**2)
    return l

result = test1(11)
print(result)
```

```python
# Now let's we have given a very large number to function

result = test1(60000000)
print(result)
```

```python
"""
We are not doing any heavy operation here just square but still we are not able to get the
result.  
"""

"""
It is not able to give the result because we are appending the result inside a single list
means we are creating a 1 memory address and try to append the data associated with the same
memory location and it will not release the memory and computation power untill the complete
calculation is not completed therefore it is generating the issue.
"""

"""
Now to solve the above kind of problem there is something called as generators.
"""
```

```python
# Generators ( yield )

def test2(n):
    for i in range(n):
        yield i**2
```

```python
test2(10)
```

```python
# While calling the above function it is returning the generator object

# So what is generator?

"""
Generator is a function that returns an iterator that produces a sequence of values when
iterated over. Generators are useful when we want to produce a large sequence of values,
but we don't want to store all of them in memory at once.
"""

"""
So generator will memorize what i have done last time what i have to do next time
and it does not need to memorize the entire data
"""
```

```python
# So here let's extract the data from the generator object.

for i in test2(10):
    print(i)
```
